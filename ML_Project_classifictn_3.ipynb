```python
x=df.drop(columns='Class',axis=1)
y=df['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_trn_prdctn=model.predict(x_train)
print(x_trn_prdctn)
```

```python
x_tst_prdctn=model.predict(x_test)
print(x_tst_prdctn)
```

```python
x_trn_accuracy=accuracy_score(y_train,x_trn_prdctn)
print(x_trn_accuracy)
```

```python
x_tst_accuracy=accuracy_score(y_test,x_tst_prdctn)
print(x_tst_accuracy)
```
