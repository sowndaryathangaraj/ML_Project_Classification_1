```python
df['Survived'].fillna(df['Survived'].mean(),inplace=True)
df['Pclass'].fillna(df['Pclass'].mean(),inplace=True)
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['SibSp'].fillna(df['SibSp'].mean(),inplace=True)
df['Parch'].fillna(df['Parch'].mean(),inplace=True)
df['Fare'].fillna(df['Fare'].mean(),inplace=True)
```

```python
df['Name'].fillna(df['Name'].mode()[0],inplace=True)
df['Sex'].fillna(df['Sex'].mode()[0],inplace=True)
df['Ticket'].fillna(df['Ticket'].mode()[0],inplace=True)
df['Cabin'].fillna(df['Cabin'].mode()[0],inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
le=LabelEncoder()
```

```python
df['Name']=le.fit_transform(df['Name'])
df['Sex']=le.fit_transform(df['Sex'])
df['Ticket']=le.fit_transform(df['Ticket'])
df['Cabin']=le.fit_transform(df['Cabin'])
df['Embarked']=le.fit_transform(df['Embarked'])
```

```python
df
```

```python
df['Survived']=df['Survived'].astype(int)
df['Pclass']=df['Pclass'].astype(int)
df['Age']=df['Age'].astype(int)
df['SibSp']=df['SibSp'].astype(int)
df['Parch']=df['Parch'].astype(int)
df['Fare']=df['Fare'].astype(int)
```

```python
df['Survived'].value_counts()
```

```python
df
```

```python
x=df.drop(columns=['PassengerId','Survived'],axis=1)
y=df['Survived']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prdctn=model.predict(x_train)
print(x_train_prdctn)
```

```python
x_test_prdctn=model.predict(x_test)
print(x_test_prdctn)
```

```python
accuracy_xtrain=accuracy_score(y_train,x_train_prdctn)
print(accuracy_xtrain)
```

```python
accuracy_xtest=accuracy_score(y_test,x_test_prdctn)
print(accuracy_xtest)
```
